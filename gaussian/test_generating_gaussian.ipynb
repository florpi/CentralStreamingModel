```python
from CentralStreamingModel.utils.read_probabilities import VD
import gsm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# First box
boxsize = 2000
snapshot = 20
box = 1


halos = VD(box, boxsize, snapshot)
```

```python
mean, std_r, std_t = gsm.interpolate_parameters(halos.r,
                    halos.mean.r, halos.std.r, halos.std.t)
```

```python
r = np.linspace(0.5,99.5, 200)

plt.plot(r, mean(r))
plt.plot(halos.r, halos.mean.r, linestyle='', marker='o',
        markersize=3)
```

```python
plt.plot(r, std_r(r))
plt.plot(halos.r, halos.std.r, linestyle='', marker='o',
        markersize=3)
```

```python
plt.plot(r, std_t(r))
plt.plot(halos.r, halos.std.t, linestyle='', marker='o',

         markersize=3)
```

```python
v_los = np.linspace(-20,20,100)
r_perp = 0.5
r_parallel = 5.5
pdf = gsm.gaussian_los_pdf( mean, std_r, std_t)
```

```python
plt.semilogy(v_los, pdf(v_los, r_perp, r_parallel))
```

```python
print(abs(v_los[1] - v_los[0]) * np.sum(pdf(v_los, r_perp, r_parallel) )) # check it's normalized
```
