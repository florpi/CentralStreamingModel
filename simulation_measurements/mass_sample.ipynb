## Test whether P(v|r,M1,M2) \aprox sqrt( P(v|r,M1,M1) * P(v|r,M2,M2)) 

```python
mean, std = compute_mean_error(central_halos, 'marginal', 'r')
```

```python
for i in range(len(mass_bins_c)):
    for j in range(i+1, len(mass_bins_c)):
        sqrt_marginal = np.sqrt(mean[:,i,i,:] * mean[:, j, j,:])

        norm = central_halos[0].wvt * np.sum(sqrt_marginal,axis=-1)

        #sqrt_marginal /= norm[:, np.newaxis] 
        
        mean_stat, std_stat = compute_mean_error(central_halos, 'mean', 'r')
        
        sqrt_approx = -np.sqrt(mean_stat[:,i,i] * mean_stat[:,j,j])
        #mu = central_halos[0].wvt * np.sum(sqrt_marginal * central_halos[0].v.t, 
        #                            axis = -1)
        plt.title(f'{mass_bins[i]:.1E} $< M_1 < ${mass_bins[i+1]:.1E},   '\
                    f'{mass_bins[j]:.1E}$ < M_2 <$ {mass_bins[j+1]:.1E}')
        
        
        plt.errorbar(central_halos[0].r, mean_stat[:,i,j], yerr= std_stat[:,i,j],
                         label='Cross measured', color='black')
        plt.plot(central_halos[0].r, mean_stat[:,i,i],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, mean_stat[:,j,j],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, sqrt_approx,
                         label='Sqrt approx')
        #plt.plot(central_halos[0].r, mu,
        #                 label='Sqrt approx PDF')
        plt.legend(bbox_to_anchor = [1,1])
        plt.show()
        
        mean_stat, std_stat = compute_mean_error(central_halos, 'std', 'r')
        
        sqrt_approx = np.sqrt(mean_stat[:,i,i] * mean_stat[:,j,j])
        #mu = central_halos[0].wvt * np.sum(sqrt_marginal * central_halos[0].v.t, 
        #                            axis = -1)

        
        
        plt.errorbar(central_halos[0].r, mean_stat[:,i,j], yerr= std_stat[:,i,j],
                         label='Cross measured', color='black')
        plt.plot(central_halos[0].r, mean_stat[:,i,i],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, mean_stat[:,j,j],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, sqrt_approx,
                         label='Sqrt approx')
        #plt.plot(central_halos[0].r, mu,
        #                 label='Sqrt approx PDF')
        plt.legend(bbox_to_anchor = [1,1])
        plt.show()
        
        mean_stat, std_stat = compute_mean_error(central_halos, 'std', 't')
        
        sqrt_approx = np.sqrt(mean_stat[:,i,i] * mean_stat[:,j,j])
        #mu = central_halos[0].wvt * np.sum(sqrt_marginal * central_halos[0].v.t, 
        #                            axis = -1)

        
        
        plt.errorbar(central_halos[0].r, mean_stat[:,i,j], yerr= std_stat[:,i,j],
                         label='Cross measured', color='black')
        plt.plot(central_halos[0].r, mean_stat[:,i,i],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, mean_stat[:,j,j],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, sqrt_approx,
                         label='Sqrt approx')
        #plt.plot(central_halos[0].r, mu,
        #                 label='Sqrt approx PDF')
        plt.legend(bbox_to_anchor = [1,1])
        plt.show()
        
        mean_stat, std_stat = compute_mean_error(central_halos, 'skewness', 'r')
        
        sqrt_approx = -np.sqrt(mean_stat[:,i,i] * mean_stat[:,j,j])
        #mu = central_halos[0].wvt * np.sum(sqrt_marginal * central_halos[0].v.t, 
        #                            axis = -1)

        
        
        plt.errorbar(central_halos[0].r, mean_stat[:,i,j], yerr= std_stat[:,i,j],
                         label='Cross measured', color='black')
        plt.plot(central_halos[0].r, mean_stat[:,i,i],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, mean_stat[:,j,j],
                         color='gray', linestyle='--')
        plt.plot(central_halos[0].r, sqrt_approx,
                         label='Sqrt approx')
        #plt.plot(central_halos[0].r, mu,
        #                 label='Sqrt approx PDF')
        plt.legend(bbox_to_anchor = [1,1])
        plt.show()
```
