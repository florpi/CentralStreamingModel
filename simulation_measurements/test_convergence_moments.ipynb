```python
import matplotlib.pyplot as plt
import numpy as np
from CentralStreamingModel.utils.read_probabilities import VD
```

```python
boxsize = 2000
box = 1
snapshot = 20

measured_v20 = VD(box, boxsize, snapshot)
```

```python
boxsize = '2000_v30'
measured_v30 = VD(box, boxsize, snapshot)
```

```python
boxsize = '2000_v40'
measured_v40 = VD(box, boxsize, snapshot)
```

```python
measured = [measured_v20, measured_v30, measured_v40]
```

```python
label = [20,30,40]
for i in range(len(measured)):
    plt.plot(measured[i].mean.r, label = label[i])
plt.legend()
```

```python
for i in range(len(measured)):
    plt.plot(measured[i].std.r, label = label[i])
plt.legend()
```

```python
for i in range(len(measured)):
    plt.plot(measured[i].std.t, label = label[i])
plt.legend()
```

```python
for i in range(len(measured)):
    plt.plot(measured[i].skewness.r, label = label[i])
plt.legend()
```

```python
for i in range(len(measured)):
    plt.plot(measured[i].kurtosis.r, label = label[i])
plt.legend()
```

```python
for i in range(len(measured)):
    plt.plot(measured[i].kurtosis.t, label = label[i])
plt.legend()
```
