```python
from models import RedshiftSpace, Streaming
import models as md
import pickle
import matplotlib.pyplot as plt
import numpy as np
import importlib
import seaborn as sns
from CentralStreamingModel.utils import plot_tools as pt
sns.set_context('talk')
import time
```

```python
box = 1
boxsize = 2000
snapshot = 20

rs = RedshiftSpace(box, 2000, 20)
```

## changing limit

```python
rs_vlim1 = RedshiftSpace(box, 2000, 20, extra='vlim1')
```

```python
rs_vlim2 = RedshiftSpace(box, 2000, 20, extra='vlim2')
```

```python
streaming_vlim0 = Streaming(rs.measured.v_los, rs.measured.r_perp,
            rs.measured.r_parallel, rs.measured.jointpdf_los, 
            rs.measured.tpcf_dict)
```

```python
streaming_vlim1 = Streaming(rs_vlim1.measured.v_los, rs_vlim1.measured.r_perp,
            rs_vlim1.measured.r_parallel, rs_vlim1.measured.jointpdf_los, 
            rs_vlim1.measured.tpcf_dict)
```

```python
streaming_vlim2 = Streaming(rs_vlim2.measured.v_los, rs_vlim2.measured.r_perp,
            rs_vlim2.measured.r_parallel, rs_vlim2.measured.jointpdf_los, 
            rs_vlim2.measured.tpcf_dict)
```

```python
s_c = streaming_vlim0.s_c
```

```python
plt.plot(s_c, s_c*s_c*rs.measured.monopole, marker='o',
        markersize=3, linestyle='', label='Direct measurement')
plt.plot(s_c, s_c*s_c*streaming_vlim0.monopole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-20,20)')
plt.plot(s_c, s_c*s_c*streaming_vlim1.monopole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-30,30)')
plt.plot(s_c, s_c*s_c*streaming_vlim2.monopole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-10,10)')

plt.legend()
```

```python
plt.plot(s_c, s_c*s_c*rs.measured.quadrupole, marker='o',
        markersize=3, linestyle='', label='Direct measurement')
plt.plot(s_c, s_c*s_c*streaming_vlim0.quadrupole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-20,20)')
plt.plot(s_c, s_c*s_c*streaming_vlim1.quadrupole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-30,30)')
plt.plot(s_c, s_c*s_c*streaming_vlim2.quadrupole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-10,10)')

plt.legend()
```

```python
plt.plot(s_c, s_c*s_c*rs.measured.hexadecapole, marker='o',
        markersize=3, linestyle='', label='Direct measurement')
plt.plot(s_c, s_c*s_c*streaming_vlim0.hexadecapole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-20,20)')
plt.plot(s_c, s_c*s_c*streaming_vlim1.hexadecapole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-30,30)')
plt.plot(s_c, s_c*s_c*streaming_vlim2.hexadecapole, marker='o',
        markersize=3, linestyle='', label = '$v_\parallel \in $ (-10,10)')

plt.legend()
```
